```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv("/Users/william/Desktop/Covid Project/Covid Data.csv")
df
```

```python

df = df.drop('USMER', axis=1)
df = df.drop('MEDICAL_UNIT', axis = 1)
df = df.drop('CLASIFFICATION_FINAL', axis = 1)
#df = df.head(10000)
values = [97,98,99]
#df = df[df.MEDICAL_UNIT.isin(values) == False]#delete
df = df[df.SEX.isin(values) == False]
df = df[df.PATIENT_TYPE.isin(values) == False]
df = df[df.INTUBED.isin(values) == False]
df = df[df.PNEUMONIA.isin(values) == False]
df = df[df.AGE.isin(values) == False]
df = df[df.PREGNANT.isin(values) == False]
df = df[df.DIABETES.isin(values) == False]
df = df[df.COPD.isin(values) == False]
df = df[df.ASTHMA.isin(values) == False]
df = df[df.INMSUPR.isin(values) == False]
df = df[df.HIPERTENSION.isin(values) == False]
df = df[df.OTHER_DISEASE.isin(values) == False]
df = df[df.CARDIOVASCULAR.isin(values) == False]
df = df[df.OBESITY.isin(values) == False]
df = df[df.RENAL_CHRONIC.isin(values) == False]
df = df[df.TOBACCO.isin(values) == False]
#df = df[df.CLASIFFICATION_FINAL.isin(values) == False] #delete
df = df[df.ICU.isin(values) == False]
#df = df[df.USMER.isin(values) == False]#delete

for col in df.columns:
    print(col)
    print(df.loc[df[col] == (97 or 98 or 99), col].count())
    print()
```

```python
df.loc[df['DATE_DIED'] != '9999-99-99', 'DATE_DIED'] = 1 #replace 
df.loc[df['DATE_DIED'] == '9999-99-99', 'DATE_DIED'] = 0 #replace
lastcol = df.pop('DATE_DIED')
df.insert(17,'DATE_DIED',lastcol)
```
