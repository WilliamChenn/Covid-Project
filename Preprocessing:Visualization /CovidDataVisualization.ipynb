```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
%matplotlib inline

df = pd.read_csv("/Users/william/Desktop/Covid Project/Cleaned_Data2.csv")
#df = pd.read_csv("/Users/william/Desktop/Covid Project/Covid Data.csv")
# Drop first column of dataframe using pop()
#df.pop(df.columns[0])
#print(df)
ax = sns.countplot(df.INTUBED)
plt.bar_label(ax.containers[0])
plt.title("Count of Intubed Missing Values",color="black");

ax = sns.countplot(df.ICU)
plt.bar_label(ax.containers[0])
plt.title("Count of ICU Missing Values",color="black");
```

```python
plt.figure(figsize=(6, 7))
sns.boxplot(data=df, x='DATE_DIED', y='AGE', palette=sns.color_palette(['#79BAEA', '#E495F7']))
plt.title('Death vs. Age Correlation', color='black', fontsize=20)
plt.xlabel(None)
plt.ylabel('Age', fontsize=14)
plt.tick_params(bottom=False)
plt.xticks([0, 1], ['DIDN\'T DIE', 'DIED'], fontsize=13);
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(
    df.DATE_DIED.value_counts().values, labels=['DIDN\'T DIE', 'DIED'], autopct='%1.1f%%',
    startangle=0, textprops={'fontsize': 16}, wedgeprops={'edgecolor': '#fff'}, colors=['#79BAEA', '#CE8CFC']
)
ax.set_title('DEATH DISTRIBUTION', fontsize=20, color='black')
plt.show()
```

```python
plt.figure(figsize=(6, 7))
sns.boxplot(data=df, x='DATE_DIED', y='AGE', palette=sns.color_palette(['#79BAEA', '#E495F7']))
plt.title('Death vs. Age Correlation', color='black', fontsize=20)
plt.xlabel(None)
plt.ylabel('Age', fontsize=14)
plt.tick_params(bottom=False)
plt.xticks([0, 1], ['DIED', 'DIDN\'T DIE'], fontsize=13);
```

```python
plt.figure(figsize=(15, 9))
sns.heatmap(df.corr(), annot=True, cmap='magma');
```
